<h2>Converting the table from the Wikipedia page in a Pandas dataframe.</h2>

```python
import pandas as pd
link = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
table = pd.read_html(link)[0] # As the first table on the page is needed.
```

```python
table.head(15)
```

<h2>Dropping all the cells in the "Borough" column which have "Not assigned" as values.</h2>

```python
no_borough = table[table["Borough"]=="Not assigned"]
table.drop(no_borough.index,inplace = True)
table.reset_index(drop=True)
```

<h2> Renaming the "Neighbourhood" cells, which have "Not assigned" values, as the name of their respective Borrough.</h2>

```python
s = table.shape
for i in range(0,s[0]):
    if table["Neighbourhood"].iloc[i]=="Not assigned":
        table.replace("Not assigned",table["Borough"].iloc[i],inplace = True)
```

```python
table = table.reset_index(drop=True)
```

```python
table.head(20)
```

```python
table.shape
```

<h2>Grouping by the "Postcode" and adding different "Neighbourhood" into the same row of a postal code.</h2>

```python
table_new = table.groupby(["Postcode","Borough"],axis = 0)["Neighbourhood"].apply(lambda x: ','.join(x))
table_new  = pd.DataFrame(table_new)
table_new.reset_index(drop=False,inplace = True)
```
