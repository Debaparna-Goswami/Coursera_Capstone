#### Converting the table from the Wikipedia page in a Pandas dataframe.

```python
import pandas as pd
link = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
table = pd.read_html(link)[0]
```

```python
table.head(15)
```

```python
no_borough = table[table["Borough"]=="Not assigned"]
table.drop(no_borough.index,inplace = True)
table.reset_index(drop=True)
```

```python
s = table.shape
for i in range(0,s[0]):
    if table["Neighbourhood"].iloc[i]=="Not assigned":
        table.replace("Not assigned",table["Borough"].iloc[i],inplace = True)
        
```

```python
table.reset_index(drop=True,inplace = True)
```

```python
table.head(20)
```

```python
table.shape
```

```python
table_new = table.groupby(["Postcode","Borough"],axis = 0)["Neighbourhood"].apply(lambda x: ','.join(x))
table_new  = pd.DataFrame(table_new)
table_new.reset_index(drop=False,inplace = True)
```

#### Merging both the dataframes to get the desired dataframe which has "Postcode","Borough","Neighbourhood","Latitude" and "Longitude".

```python
table_with_location = pd.merge(table_new,loc_data,how = "outer")
table_with_location.rename(columns = {"Postcode":"PostalCode"},inplace = True)
table_with_location.head()
```

#### We need the location data of Toronto to segment and cluster the neighbourhood.

```python
table_borough = table_with_location[["Borough"]]
toronto_table = table_with_location[table_borough.apply(lambda row: row.astype(str).str.contains('Toronto').any(), axis=1)]
toronto_table
```

#### Getting the Latitude and Longitude of Toronto, Canada - 

```python
address = 'Toronto, CANADA'

geolocator = Nominatim(user_agent="Toronto_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

#### Creating a map of Toronto using Folium library

```python
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)
```

#### Superimposing the neighbourhoods obtained from the dataframe <b>"toronto_table"</b> on the <b>"map_toronto"</b>-

```python
for lat, lng, borough, neighbourhood in zip(toronto_table['Latitude'], toronto_table['Longitude'], toronto_table['Borough'], toronto_table['Neighbourhood']):
    label = '{}, {}'.format(neighbourhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='green',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
